```python
import os
import pickle
from sentence_transformers import SentenceTransformer
import chromadb

# Initialize the ChromaDB client
client = chromadb.Client()

folder_path = r"C:\Users\MCOB PHD 14\Dropbox\Charlie's Dissertation\Beige Books\chunks"


# Initialize the sentence-transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Read text files and generate embeddings
texts = []
document_names = []
for file_name in os.listdir(folder_path):
    if file_name.endswith('.txt'):
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
            texts.append(text)
            document_names.append(file_name)

# Generate embeddings for the documents
embeddings = model.encode(texts, show_progress_bar=True)

# Create a ChromaDB collection and add documents
collection = client.create_collection("beige_books_embeddings")

# Generate unique IDs for each document
document_ids = [f"doc_{i}" for i in range(len(document_names))]

# Add documents to the collection
collection.add(
    ids=document_ids,
    documents=texts,
    embeddings=embeddings,
    metadatas=[{"filename": name} for name in document_names]
)

# Save embeddings and metadata to disk
save_path = r'C:\Users\MCOB PHD 14\Dropbox\Charlies Dissertation\Beige Books\chunks\saved_data'
os.makedirs(os.path.dirname(save_path), exist_ok=True)

with open(f'{save_path}_embeddings.pkl', 'wb') as f:
    pickle.dump(embeddings, f)

with open(f'{save_path}_document_names.pkl', 'wb') as f:
    pickle.dump(document_names, f)

with open(f'{save_path}_texts.pkl', 'wb') as f:
    pickle.dump(texts, f)

print("Data processed and saved successfully.")
```

```python
import pickle
from sentence_transformers import SentenceTransformer
import chromadb

# Load embeddings and metadata
load_path = r'C:\Users\MCOB PHD 14\Dropbox\Charlies Dissertation\Beige Books\chunks\saved_data'

with open(f'{load_path}_embeddings.pkl', 'rb') as f:
    embeddings = pickle.load(f)

with open(f'{load_path}_document_names.pkl', 'rb') as f:
    document_names = pickle.load(f)

with open(f'{load_path}_texts.pkl', 'rb') as f:
    texts = pickle.load(f)

# Initialize the ChromaDB client and access the collection
client = chromadb.Client()
collection = client.get_collection("beige_books_embeddings")

# Generate unique IDs for each document
document_ids = [f"doc_{i}" for i in range(len(document_names))]

# Add the loaded data to the collection
collection.add(
    ids=document_ids,
    documents=texts,
    embeddings=embeddings,
    metadatas=[{"filename": name} for name in document_names]
)

# Initialize the sentence-transformer model for generating query embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Search the collection
query = "negative outlook"  # Replace with your search query
query_embedding = model.encode([query])

# Perform the search
results = collection.query(
    query_embeddings=query_embedding,
    n_results=5  # Number of similar documents to retrieve
)

# Iterate through the results and print the output
for i in range(len(results['documents'][0])):  # Accessing the first list inside 'documents'
    print(f"Result {i + 1}:")
    print("Document:", results['metadatas'][0][i]['filename'])  # Accessing the filename
    print("Score:", results['distances'][0][i])  # Accessing the similarity score
    print("Snippet:", results['documents'][0][i][:500])  # Displaying a snippet of the document content
    print()
```
